```python
import os

from Bio import SearchIO
```

# creating custom blast db

```python
blastn_db = "makeblastdb -in chr1.fa -out chr1 -parse_seqids -dbtype nucl"
os.system(blastn_db)
```

 -task "blastn-short: allows BLAST to work with shorter (20nt ish) sequences

```python
short_blast = "blastn -db chr1 -query test.fa -outfmt 6 -out result.tsv -task blastn-short"

os.system(short_blast)
```
