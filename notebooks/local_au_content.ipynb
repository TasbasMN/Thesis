we have at least 2k results that we can't get flanking dinucleotides.

Also, flanking dinucleotides is a context within canonical matches so we can't reliably find flanking dinucleotides of non-canonical sites.

For example, an alignment string (CLASH index: 18490)

1111111111011110111110

where do we draw the line as the "interface" ?


00000111101010100000000000000000

if we draw the lines of interface as following;

xx(111111111101111011111)0x

then we have 2 unknown nucleotides at the 5' flank because CLASH doesn't supply that information.

we can't generate them because we would introduce AU content bias. If we generate a sequence with neutral (0.5) AU content to fill there, the *actual* AU content might be different.
