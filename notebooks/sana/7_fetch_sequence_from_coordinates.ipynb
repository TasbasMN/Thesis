```python
import os
from pyensembl import EnsemblRelease
import pandas as pd

pd.set_option("display.max_columns", None)


# set pyensembl cache location
os.environ['PYENSEMBL_CACHE_DIR'] = "../data"

# init database
g37 = EnsemblRelease(75)

# download and index db
g37.download()
g37.index()
```

```python
df = pd.read_csv("data/external/sana/final_dataset.csv")
uniques = df.SYMBOL.unique().tolist()

def write_list_to_file(lst, filename):
    with open(filename, 'w') as file:
        for item in lst:
            file.write(str(item) + '\n')

write_list_to_file(uniques, "uniques.txt")
```

```python
def get_gene_names_at_locus(contig, position):
    return g37.gene_names_at_locus(contig, position)

def get_gene_ids_at_locus(contig, position):
    return g37.genes_at_locus(contig, position)


df["gene_name"] = df.apply(lambda row: get_gene_names_at_locus(row["chr"], row["start"]), axis=1)
df["gene_id"] = df.apply(lambda row: get_gene_ids_at_locus(row["chr"], row["start"]), axis=1)
```
