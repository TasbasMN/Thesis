```python
%cd ..

import pandas as pd

```

```python
def find_seed_type(df):
    df['anchor_a'] = df['mre_region'].apply(lambda x: 1 if x[-1] == 'A' else 0)

    df['seed_match'] = df['alignment_string'].apply(
        lambda x: 1 if x[1:8].count('0') == 0 else 0)

    df['seed_1_mismatch'] = df['alignment_string'].apply(
        lambda x: 1 if x[1:8].count('0') == 1 else 0)

    return df

def add_ta_sps_columns(df):
    # generating seed col
    df["seed"] = df["mirna_sequence"].apply(lambda x: x[1:8]).str.replace('T', 'U')
    
    # reading ta sps data
    ta_sps_df = pd.read_csv("data/processed/ta_sps/ta_sps.csv", usecols=["seed_8mer", "ta_log10", "ta_percentile_rank", "sps_mean", "sps_mean_percentile_rank"])
    ta_sps_df = ta_sps_df.rename(columns={"seed_8mer": "seed"})
    
    # merge
    df = pd.merge(df, ta_sps_df, on="seed", how="left")
    
    # drop temp col
    df.drop(columns=["seed"], inplace=True)
    
    return df
```

```python
df = pd.read_csv('results/merged_df_results.csv')
df = find_seed_type(df)
df = add_ta_sps_columns(df)


df.to_csv("results/final.csv", index=False)
df.head()
```
