```python
def placeholder_top(i, mutation_df):
    row = mutation_df.iloc[i]

    mrna_seq = row.mrna_sequence[row.mrna_start:row.mrna_end].replace("T", "U")
    mrna_db = row.mrna_dot_bracket_5to3.replace("(", "|")
    mirna_db = row.mirna_dot_bracket_5to3.replace(")", "|")[::-1]
    mirna_seq = row.mirna_sequence[row.mirna_start:row.mirna_end].replace("T", "U")[::-1]

    mrna_pointer = 0
    mirna_pointer = 0

    while mrna_pointer < len(mrna_seq) and mirna_pointer < len(mirna_seq):
        if mrna_db[mrna_pointer] == "." and mirna_db[mirna_pointer] == ".":
            pass

        elif mrna_db[mrna_pointer] == "|" and mirna_db[mirna_pointer] == "|":
            pass

        elif mrna_db[mrna_pointer] == "|" and mirna_db[mirna_pointer] == ".":
            mrna_db = f"{mrna_db[:mrna_pointer]} {mrna_db[mrna_pointer:]}"
            mrna_seq = f"{mrna_seq[:mrna_pointer]} {mrna_seq[mrna_pointer:]}"

        elif mrna_db[mrna_pointer] == "." and mirna_db[mirna_pointer] == "|":
            mirna_db = f"{mirna_db[:mirna_pointer]} {mirna_db[mirna_pointer:]}"
            mirna_seq = f"{mirna_seq[:mirna_pointer]} {mirna_seq[mirna_pointer:]}"

        mrna_pointer += 1
        mirna_pointer += 1

    print (f"""
                {mirna_seq}
                {mirna_db}
                {mrna_db}
                {mrna_seq}
                
                
            """)

def placeholder_bottom(i, mutation_df):
    row = mutation_df.iloc[i]

    mrna_seq = row.mrna_sequence[row.mrna_start:row.mrna_end].replace("T", "U")
    mrna_db = row.mrna_dot_bracket_5to3.replace("(", "|")
    mirna_db = row.mirna_dot_bracket_5to3.replace(")", "|")[::-1]
    mirna_seq = row.mirna_sequence[row.mirna_start:row.mirna_end].replace("T", "U")[::-1]

    mrna_pointer = 0
    mirna_pointer = 0

    while mrna_pointer < len(mrna_seq) and mirna_pointer < len(mirna_seq):
        if mrna_db[mrna_pointer] == "." and mirna_db[mirna_pointer] == ".":
            pass

        elif mrna_db[mrna_pointer] == "|" and mirna_db[mirna_pointer] == "|":
            pass

        elif mrna_db[mrna_pointer] == "|" and mirna_db[mirna_pointer] == ".":
            mrna_db = f"{mrna_db[:mrna_pointer]} {mrna_db[mrna_pointer:]}"
            mrna_seq = f"{mrna_seq[:mrna_pointer]} {mrna_seq[mrna_pointer:]}"

        elif mrna_db[mrna_pointer] == "." and mirna_db[mirna_pointer] == "|":
            mirna_db = f"{mirna_db[:mirna_pointer]} {mirna_db[mirna_pointer:]}"
            mirna_seq = f"{mirna_seq[:mirna_pointer]} {mirna_seq[mirna_pointer:]}"

        mrna_pointer += 1
        mirna_pointer += 1

    print (f"""
                {mrna_seq}
                {mrna_db}
                {mirna_db}
                {mirna_seq}
            """)
            
```

```python
df = pd.read_csv("results/predictions_0_50.csv")
df =  (df.assign(len_mrna_dotbracket=lambda x: x["mrna_dot_bracket_5to3"].str.len())
         .assign(len_mirna_dotbracket=lambda x: x["mirna_dot_bracket_5to3"].str.len())
         .assign(dot_bracket_absolute_diff=lambda x: abs(x["len_mrna_dotbracket"] - x["len_mirna_dotbracket"])))

# mask to find ids with changed prediction
mask = df['id'].isin(df.groupby(pd.Grouper(key="id")).prediction.sum()[lambda x: x == 1].index)

```
